```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
import datetime

today = datetime.datetime.now()
ds,de = input().split()

cos = ['삼성','롯데','현대차','신세계','SK','LG','네이버','카카오','GS','KT','CJ','넷마블']
keywords = ['NFT','메타버스']

key_data = []
data = []

for co in cos:
    for keyword in keywords:        
        url = f"https://search.naver.com/search.naver?where=news&query={co+keyword}&sm=tab_opt&sort=0&photo=0&field=0&pd=3&ds={ds}&de={de}&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom20220409to20220410&is_sug_officeid=0"
        res = requests.get(url)
        res.raise_for_status()

        soup = BeautifulSoup(res.text,'lxml')
        titles = soup.find_all('a',attrs={'class':'news_tit'})

        for title in titles:
            if keyword in title.text:
                key_data.append([f'{co} {keyword}',title.text,title['href']])
            else:
                data.append([f'{co} {keyword}',title.text,title['href']])
df1 = pd.DataFrame(key_data,columns=['검색어','제목','URL'])
df2 = pd.DataFrame(data,columns=['검색어','제목','URL'])
```
