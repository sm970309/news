```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time
import pandas as pd
import re
import googletrans
```

```python
browser = webdriver.Chrome()
browser.maximize_window()
url = 'https://www.soompi.com/latest'
browser.get(url)
for _ in range(3):
    browser.find_element_by_xpath('//*[@id="app"]/div[1]/div[4]/div/div[2]/a').click()
    time.sleep(0.5)
```

```python
soup = BeautifulSoup(browser.page_source,"lxml")
browser.close()
data = soup.find_all('div',attrs={'class':'thumbnail-wrapper'})
articles = []
for d in data:
    src = d.find('a')['href']
    articles.append(f'https://www.soompi.com{src}')
```

```python
news=[]
for i,article in enumerate(articles):
    tmp = {}
    res = requests.get(article)
    soup = BeautifulSoup(res.text,'lxml')
    thumb = soup.find('span',attrs={'class':'image-wrapper'})
    tmp_txt=[str(thumb).replace('data-src','src')]
    title = soup.find('div',attrs={'class':'article-info'})
    tmp['title'] = title.find('h1').text
    tmp['domain'] = 'www.soompi.com'
    no = re.findall('[0-9]*wpp',article)
    tmp['no'] = no[0].replace('wpp','')
    content = soup.find('div',attrs={'class':'article-wrapper'})
    body = content.findChild()
    if body.find('blockquote'):
        continue
    if body.find('span',attrs={'class':'embed-youtube'}):
        continue
    for child in body.children:
        if "Source" in child.text:
            break
        elif "WatchNow" in child.text.replace(' ',''):
            tmp_txt = tmp_txt[:-2]
            break
        if child.text==' ':
            continue
        tmp_txt.append(str(child).replace('data-src','src').replace("\xa0", " "))
    tmp['content'] = '<br>'.join(tmp_txt)
    news.append(tmp)
```

```python
for n in news:
    print(n['content'],end='\n\n')
```

```python
df = pd.json_normalize(news)
df.to_excel('20220621.xlsx')
```
