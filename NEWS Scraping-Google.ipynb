```python
import feedparser
import pandas as pd
import datetime

today = datetime.datetime.now()
date = (f'{today.year}{str(today.month).zfill(2)}{str(today.day).zfill(2)}')
```

```python
f = feedparser.parse('https://news.google.com/rss/search?q=bayc%20NFT%20when%3A1d&hl=en-US&gl=US&ceid=US%3Aen')
data = []
for e in f.entries:
    data.append([e.published,e.title,e.link])
df = pd.DataFrame(data,columns=['날짜','제목','URL'])
```

```python
df = df.sort_values(['날짜'])
df
```

```python
with pd.ExcelWriter(f'bayc{date}.xlsx') as w:
    df.to_excel(w,sheet_name='bayc')
```
